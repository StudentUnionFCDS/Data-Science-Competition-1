Dealing with train categorical data

```python
categorical_df=train.Genre
categorical=pd.get_dummies(categorical_df,drop_first=True)
categorical.head()
```

```python
train.drop(['CustomerID','Genre'],1,inplace=True)
train.head()
```

```python
train_final=pd.concat([train,categorical],1)
train_final.head()
```

Dealing with test categorical data

```python
categorical_test=test.Genre
cate=pd.get_dummies(categorical_test,drop_first=True)
cate.head()
```

```python
test.drop(['CustomerID','Genre'],1,inplace=True)
test.head()
```

```python
test_final=pd.concat([test,cate],1)
test_final.head()
```

```python
X_train=train_final.drop('Class',1)
y_train=train_final.Class
```

Importing random forest classifier which doesnt need feature scalling

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
model.fit(X_train,y_train)
```

```python
Y_pred_test = model.predict(test_final)
Y_pred_test
```
